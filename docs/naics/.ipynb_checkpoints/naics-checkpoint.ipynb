```r
### Demonstration of NAICS Code Use
### kbmorales@protonmail.com

# This notebook gives some brief examples of the use of the PPP loan data given a little 
# more detail on industry types with NAICS codes labels.

# setup
suppressWarnings(library(tidyverse))
options(repr.plot.res = 300)
```

```r
# load csv -- prepared separately
# LARGE FILE - give it a minute
adbs = read_csv(here::here("data","tidy_data","adbs_naics.csv"),
               col_types = "cccccccccccccccccccccccccl")
# note: providing col specifications in shorthand due to large amount of NAs 
```

```r
# glimpse
glimpse(adbs)
```

```r
# Examples of column specifications for NAICS:
adbs  %>% 
    select(starts_with("naics"))  %>% 
    sample_n(10)
```

```r
# Look at all top-level NAICS industries in data

# NOTE: when unable to exactly match the NAICS industry code, I used the first 2 digits to at least match with
# the highest level in the NAICS heirarchy.

# After that:
# NAICS Code is missing from 133.5k entries, and is "999990" for 88.5k entries (appears to be 'unknown') 
# This is 4.5% of all loans

adbs %>% 
  count(naics_lvl_1) %>% 
  ggplot(aes(x = reorder(naics_lvl_1,n),
             y = n)) +
  geom_col() +
  coord_flip() +
  labs(x = "",
       y = "Count") +
  theme_minimal()
```

```r
# Loan range distribution for each NAICS level 1
adbs %>% 
  group_by(naics_lvl_1) %>% 
  filter(!str_detect(LoanRange_Unified,
                    "^l|^m|^n|^o|^p")) %>% 
  count(LoanRange_Unified) %>% 
  na.omit() %>% 
  ggplot(aes(x = LoanRange_Unified,
             y = n,
             fill = LoanRange_Unified)) +
  geom_col() +
  facet_wrap(~naics_lvl_1,
             nrow = 5) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1),
        legend.position = 'none') +
  labs(x = "Loan Range",
       y = "# of Loans") +
  scale_fill_viridis_d()
```

```r
# Use case example: filter to all healthcare-related PPP loans in Maryland
health_md = adbs  %>% 
    filter(State == "MD",
           naics_lvl_1 == "Health Care and Social Assistance"
          )

sample_n(health_md, 10) # sample 10
```

```r
# Further refine filters with additional NAICS levels: e.g., hospitals
health_md %>% 
    filter(naics_lvl_2 == "Hospitals")  %>% 
    sample_n(10) # sample 10
```

```r
# Interesting Note: the NAICS codes that are missing aren't evently distributed.
# May 1 and May 3, may be interesting to look at those loans specifically.

adbs %>% 
  filter(is.na(naics_lvl_1)) %>% 
  mutate(DateApproved = as.Date(DateApproved,
                                format="%m/%d/%Y")) %>% 
  group_by(DateApproved) %>% 
  count(NAICSCode) %>% 
  # count(DateApproved) %>% 
  ggplot(aes(x = DateApproved,
             y = n,
             fill = NAICSCode)) +
  geom_area() +
  theme_minimal() +
  scale_x_date(date_breaks = "week",
               date_minor_breaks = "day") +
  theme(legend.position = c(0.8,0.8),
        axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)
        ) +
  labs(x = "",
       y = "# of Loans",
       title = "Loans with a missing or 999990 NAICS Code")
```
